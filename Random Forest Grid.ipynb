```python
#importing libraries-pandas(It contains functions to perform data base operation like reading a file etc..)
import pandas as pd
```

```python
data=pd.read_csv("insurance_pre.csv")
data
```

```python
data=pd.get_dummies(data,drop_first=True)
data
```

```python
data.columns
```

```python
independent=data[["age", "bmi", "children", "sex_male","smoker_yes"]]
dependent=data[["charges"]]
```

```python
#Without spliting the data as 
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
#Getting the parameters in the Dictionary
param_grid={'criterion':['squared_error','friedman_mse','absolute_error','poisson'],'max_features':['sqrt','log2'],'n_estimators':[1,10,100,1000]}
grid=GridSearchCV(RandomForestRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_
print("The Rsquare value for the the best paramater {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("bmi:"))
children_input=float(input("children:"))
sex_input=int(input("Sex Male 0 or 1:"))
smoker_input=int(input("smoke yes 0 or 1:"))
```

```python
future_predictions=grid.predict([[age_input,bmi_input,children_input,sex_input,smoker_input]])
print("The future prediction is : {}".format(future_predictions))
```
